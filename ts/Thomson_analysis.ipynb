```python
#filter data for fit
radius_idx = 140
xdata = wavelength[radius_idx,:]
ydata = intensity[radius_idx,:]

#test filter
plt.axvline(x=6000,color="red")
plt.axvline(x=6400,color="red")
plt.axvline(x=6700,color="red")
plt.axvline(x=6850,color="red")
plt.axvline(x=7100,color="red")
plt.axvline(x=7500,color="red")

plt.plot(xdata,ydata) 
plt.xlabel("wavelength (Angstrom)")
plt.ylabel("intensity (au)")
plt.title("intensity spectrum @r = "+str(radius[radius_idx])+"m (close to middle)")
plt.xlim(6e3,8e3)
plt.show()
```

```python
def filter_data(xdata,ydata):
    xdata2, ydata2 = [],[]
    for i in range(len(xdata)):
        Append = False
        if xdata[i] > 6200 and xdata[i] < 6400:
            Append = True
        if xdata[i] > 6700 and xdata[i] < 6850:
            Append = True
        if xdata[i] > 7100 and xdata[i] < 7500:
            Append = True
        if Append:
            xdata2.append(xdata[i])
            ydata2.append(ydata[i])
    return xdata2, ydata2
```

```python
#plot filtered data
xdata2, ydata2 = filter_data(xdata,ydata)
plt.scatter(xdata2, ydata2,marker="x")
```

```python
#fit spectrum for one position
kb = const.k
me = const.m_e
c = const.c
radius_idx = 140

def spectrum(lambda_s,Te,A0):
    lambda_i=6943
    theta = angle[int(radius_idx)]
    eV2J = 1.602e-19
    beta_th = np.sqrt(2*Te*eV2J/(me*c**2)) #omit boltzmann const. and insert eV to J -> T in eV
    sigma_lambda = lambda_i*beta_th*np.sqrt(2)*np.sin(theta/2)
    
    intensity = A0*np.exp(-0.5*((lambda_s-lambda_i)/sigma_lambda)**2)
    return intensity


popt, cov = curve_fit(spectrum,xdata2,ydata2,p0=[1e3,2e3])

print("Te = "+str(popt[0])+"eV @r= "+str(radius[radius_idx])+"m")
```

```python
#look at fit result
plt.scatter(xdata2, ydata2,marker="x")

x = np.linspace(min(xdata),max(xdata),1000)
plt.plot(x,spectrum(x,popt[0],popt[1]))

plt.xlabel("wavelength (Angstrom)")
plt.ylabel("intensity (au)")
plt.title("intensity spectrum @r = "+str(radius[radius_idx])+"m (close to middle)")
plt.show()
```

```python
maxFit = np.argmax(spectrum(x,popt[0],popt[1]))
print("fitted gaussian has peak at: ",x[maxFit])
```

**Note:** As explained earlier at the end of section 3, we see that the peak of the gaussian is baiscally at the wavelength of the laser, even though this wavelength was filtered out. Hence, fits matches expectation.

### Now do fit for all positions

estimate photon count from emission by using result from a.)

```python
print("Remember: Number of photons collected on lens: ",Ncol)

intensity_sum = np.sum(intensity)
I2photons = Ncol/intensity_sum #units: photons/intensity Ncol: Number of photons collected on lens

print("Hence assuming no losses, number of photons causing unit intensity: ",I2photons)
```

### estimate the Poisson error
Calculate by using $\epsilon = \frac{Intensity[counts]}{\sqrt{photonsEvents}}$ where $photonEvents = I2photons[photons/counts]*Intensity[counts]$ with $I2photons$ is conversion constant calculated above.

```python
#cut beginning and end data
front_cut = 15
back_cut = 30
radius2 = radius[front_cut:-back_cut]

Te = np.zeros(len(radius2))
Te_errors = np.zeros(len(radius2))

for i in range(len(radius2)): 
    radius_idx = i+front_cut
    xdata = wavelength[radius_idx,:]
    ydata = intensity[radius_idx,:]
    xdata2, ydata2 = filter_data(xdata,ydata)
    
    poisson_error = np.ones(len(ydata2)) #sigma = 1 is default for no specified error
    for j in range(len(ydata2)):
        photon_events[j] = ydata2[j]*I2photons
        if ydata2[j] != 0:
            poisson_error[j] = ydata2[j]/np.sqrt(photon_events[j]) #this is where the error is caluclated
    
    
    popt, cov = curve_fit(spectrum,xdata2,ydata2,p0=[300,2e3],sigma=poisson_error) 
    
    Te_errors[i] = np.sqrt(cov[0,0])
    Te[i] = popt[0]
```

```python
plt.errorbar(radius2,Te, yerr=Te_errors,fmt='o',label="T_e with errors from fit")

plt.xlabel("radial position in m",fontsize=14)
plt.ylabel(r"$T_e$ in eV",fontsize=14)
plt.title("electron temperature profile for #17447 from TS",fontsize=14)
plt.legend(loc="lower center")
plt.ylim(0,500)
plt.show()
```

### consideration of errors

Sources of error in the spectrum data include: <br>
<ol>
<li>The error of the fit <i>which has been included</i> in above plot
<li>As mentioned in extension 1 of script, the spectrometer which effectively converts different wavelengths of the emitted light to different reflection angles off of the diffraction grating and hence to specific locations on the CCD array, does not do this perfectly. Hence locations on the CCD array associated with one specific wavelength will also receive some photons from other wavelengths. This can be captured by wrapping the already existing gaussian by an additional gaussian that represents the instrument function of the spectrometer.
<li>The intensity is discretely measured (or at least proportional) counts and hence has a uniform distribution of +/-1 and hence std of +/- 0.29. <i>This error is negligable howeber.</i>
<li>The viewing angle $\theta$ of each radius will have an error that effects the spectrum. 
<li>There will be noise associated with the Poisson distribution of the discrete and probabilistic nature of photon emissions. For this noise the $SNR = \sqrt{N}$ or $Noise = Signal/\sqrt{N}$ where N is the number of events. This is included in the fit by assuming no photon losses in the instrument and hence estimating the conversion from intensity to photon count.
<li>As extension 2 mentiones, the theoretical model of the spectrum can be adjusted for reltaivistic effect. Different/adjusted model will give better approximation for temperature measurement. However calculation below shows that 450 eV of pedestal corresponds to ~0.04c in thermal speeds. Relativistic correction not expected to cause large changes in estimated $T_e$ here.
</ol>
<br>
